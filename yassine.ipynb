```python
import pandas as pd 
df=pd.read_csv("data\BTCUSDT.csv")
```

```python
df["open_time"] = pd.to_datetime(df["open_time"], unit='ms', origin='unix')
df["close_time"] = pd.to_datetime(df["close_time"], unit='ms', origin='unix')
```

```python
df = df.sort_values(by='open_time', ascending=False)
```

```python
df = df.set_index('open_time')
```

```python
import matplotlib.pyplot as plt

# Configurar el tamaño del gráfico
plt.figure(figsize=(12, 6))

# Crear el gráfico de 'close' con el índice del DataFrame en el eje x
plt.plot(df2.index, df2['close'])

# Configurar los ejes y el título del gráfico
plt.xlabel('Fecha')
plt.ylabel('Close')
plt.title('Precio de Cierre')

# Rotar las etiquetas del eje x para mejorar la legibilidad
plt.xticks(rotation=45)

# Mostrar el gráfico
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_pacf, plot_acf
import matplotlib.pyplot as plt

# Obtener la columna que deseas graficar (por ejemplo, 'close')
data = df['close']

# Crear la figura y los subplots
fig = plt.figure(figsize=(15, 6))
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 2, 3)
ax3 = fig.add_subplot(2, 2, 4)

# Graficar los datos en el primer subplot
ax1.plot(data, color='black')
ax1.set_title('Serie de tiempo')

# Calcular y graficar la función de autocorrelación en el segundo subplot
plot_acf(data, lags=20, auto_ylims=True, ax=ax2)
ax2.set_title('Autocorrelación')

# Calcular y graficar la función de autocorrelación parcial en el tercer subplot
plot_pacf(data, lags=20, method='ywm', auto_ylims=True, ax=ax3, color='r')
ax3.set_title('Autocorrelación Parcial')

# Ajustar los espacios entre los subplots
fig.tight_layout()

# Mostrar el gráfico
plt.show()
```
